```python
VESIcal_out=v.BatchFile(filename=None, dataframe=df_Tput2008, label='Sample_ID')
VESIcal_out.data
```

```python
chunk_size = 20

chunks = np.array_split(VESIcal_out.data, len(VESIcal_out.data) // chunk_size)
if len(VESIcal_out.data) % chunk_size != 0:
    num_chunks=len(chunks)
    last_chunk = VESIcal_out.data[num_chunks * chunk_size:]
    chunks.append(last_chunk)
```

```python
import multiprocessing
import sys
import io
import pandas as pd
import numpy as np
from tqdm.contrib.concurrent import process_map
from tqdm import tqdm
import time

start_time = time.time()

def process_chunk(chunk, chunk_num):
    processed_chunk = pd.DataFrame()
    sys.stdout = io.StringIO()
    for i in tqdm(range(len(chunk)), desc=f'Processing Chunk {chunk_num}'):
        sub = chunk[i:i+1]
        sub_batchfile = v.BatchFile(filename=None, dataframe=sub, label=sub.index)

        try:
            sub_satP = sub_batchfile.calculate_saturation_pressure(temperature="Temp")
        except Exception as e:
            print(f"An error occurred: {e}")
            sub_satP = sub_batchfile
            sub_satP = sub_satP.assign(**{'SaturationP_bars_VESIcal': np.nan,
                                          'XH2O_fl_VESIcal': np.nan,
                                          'XCO2_fl_VESIcal': np.nan,
                                          'FluidMass_grams_VESIcal': np.nan,
                                          'FluidSystem_wt_VESIcal': np.nan,
                                          'Model': 'MagmaSat',
                                          'Warnings':'Could not converge'})
        processed_chunk = pd.concat([processed_chunk, sub_satP])
    return processed_chunk

# Define a function for multiprocessing with tqdm progress bars
def process_wrapper(chunk_num, chunk):
    return process_chunk(chunk, chunk_num)
```

```python
select_chunks = chunks[:20]
start_time = time.time()
results = process_map(process_wrapper, range(len(select_chunks)), select_chunks, desc='Processing Chunks', max_workers=multiprocessing.cpu_count())

end_time = time.time()
execution_time = end_time - start_time
print("Execution time:", execution_time, "seconds")

combined_df = pd.concat(results)
combined_df
```
