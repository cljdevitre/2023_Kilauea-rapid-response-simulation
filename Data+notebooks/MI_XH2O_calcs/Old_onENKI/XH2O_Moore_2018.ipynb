```python
import VESIcal as v
# !pip install Thermobar
import Thermobar as pt
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
out=pt.import_excel('MI_compilation_XH2O.xlsx', sheet_name="Moore_2018_kluychev")
my_input=out['my_input']
myOls=out['Ols']
myLiquids1=out['Liqs']

## Lets check the outputs have loaded right
display(myOls.head())
display(myLiquids1.head())
```

```python
T_HT87=pt.calculate_liq_only_temp(liq_comps=myLiquids1, equationT="T_Helz1987_MgO")
df_HT87=pt.convert_to_vesical(liq_comps=myLiquids1,
                              T1=T_HT87, unit='Kelvin', Fe3Fet_Liq=0.15)
df_HT87.head()
```

```python
VESIcal_out=v.BatchFile(filename=None, dataframe=df_HT87, label='Sample_ID')
VESIcal_out.data
```

```python
df= VESIcal_out.get_data()
chunk_size = 20
subframes = [v.BatchFile(filename=None,dataframe=df.iloc[i:i+chunk_size],label=None) for i in range(0, len(df), chunk_size)]
```

```python
for index, data in enumerate(subframes):
    if index>=0:
        Sat_P_MagmaSat = data.calculate_saturation_pressure(temperature="Temp")
        filename = f"XH2O_Moore_2018_kluychev_{index}.xlsx"
        Sat_P_MagmaSat.to_excel(excel_writer=filename)
```

```python
import os
combined_data = pd.DataFrame()

for index in range(len(subframes)):
    filename = f"XH2O_Moore_2018_kluychev_{index}.xlsx"
    if os.path.exists(filename):
        df = pd.read_excel(filename)
        combined_data = pd.concat([combined_data, df])

combined_filename = "XH2O_Moore_2018_kluychev_combined.xlsx"
combined_data.to_excel(combined_filename, index=False)
```
